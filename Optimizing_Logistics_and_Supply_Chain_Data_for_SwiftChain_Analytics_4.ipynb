Data Preprocessing

Encode Categorical Columns

```python
columns_to_encode = ['customer_segment', 'shipping_mode', 'order_status']

# Check if columns exist before encoding
missing_columns = [col for col in columns_to_encode if col not in data.columns]

if missing_columns:
    print(f"Error: The following columns were not found in the DataFrame: {missing_columns}")
else:
    data = pd.get_dummies(data, columns=columns_to_encode, drop_first=True)
    print("Categorical columns encoded successfully.")
    print(data.head()) # Display the head of the updated DataFrame
```

Handle Missing Values

```python
data.fillna(data.median(numeric_only=True), inplace=True)
```

Normalize Numerical Features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['profit_per_order', 'sales_per_customer']] = scaler.fit_transform(data[['profit_per_order', 'sales_per_customer']])
```

Feature Engineering

Shipping Duration

```python
data['order_date'] = pd.to_datetime(data['order_date'], errors='coerce')
data['shipping_date'] = pd.to_datetime(data['shipping_date'], errors='coerce')
data['shipping_duration'] = (data['shipping_date'] - data['order_date']).dt.days
```

Engineer features from `product_category_id` and `customer_country`.

```python
# Feature Engineering from product_category_id: Category Frequency
category_frequency = data['product_category_id'].value_counts(normalize=True)
data['category_frequency'] = data['product_category_id'].map(category_frequency)

# Feature Engineering from customer_country: Average Shipping Duration per Country
# Calculate the average shipping duration for each country, handling potential NaNs in shipping_duration
country_avg_shipping_duration = data.groupby('customer_country')['shipping_duration'].mean()

# Map the average shipping duration back to the original DataFrame
data['country_avg_shipping_duration'] = data['customer_country'].map(country_avg_shipping_duration)

print("Features engineered successfully. Displaying head of the updated DataFrame:")
display(data[['product_category_id', 'category_frequency', 'customer_country', 'country_avg_shipping_duration']].head())
```

Perform feature selection to identify the most predictive variables.

```python
from sklearn.feature_selection import SelectKBest, f_classif
import numpy as np

# Separate features (X) and target (y)
# Exclude non-numeric columns for f_classif, except the engineered numeric ones
X = data.select_dtypes(include=np.number).drop('label', axis=1)
y = data['label']

# Handle potential infinite values which can occur after scaling or feature engineering
X.replace([np.inf, -np.inf], np.nan, inplace=True)

# Impute NaNs that might have been introduced by coercing datetime errors or inf replacement
# Using median imputation as done previously
X.fillna(X.median(), inplace=True)


# Select top K features (e.g., top 20) using f_classif
# f_classif works well with numerical features and a classification target
k_best_features = SelectKBest(score_func=f_classif, k=20) # You can adjust k

fit = k_best_features.fit(X, y)

# Get scores and feature names
df_scores = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(X.columns)

# Concatenate dataframes
feature_scores = pd.concat([df_columns, df_scores], axis=1)
feature_scores.columns = ['Feature', 'Score']

# Sort features by score
feature_scores.sort_values(by='Score', ascending=False, inplace=True)

print("Top 20 features based on F-statistic:")
display(feature_scores.head(20))

# Select the top k features from the original DataFrame
selected_features = feature_scores['Feature'].head(20).tolist()
X_selected = X[selected_features]

print("\nShape of the selected features DataFrame:")
print(X_selected.shape)
```

PHASE 5: Model Development (Week 4â€“6)
Actions:
Split the dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# Using the selected features (X_selected) and the target variable (y)
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)

print("Data split successfully.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

Train and evaluate multiple classification algorithms (e.g., Logistic Regression, Random Forest, XGBoost).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable for XGBoost
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)


# Initialize models
log_reg = LogisticRegression(max_iter=5000, random_state=42) # Increased max_iter
rf_clf = RandomForestClassifier(random_state=42)
# Use the encoded target variable for XGBoost
xgb_clf = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')


models = {
    "Logistic Regression": log_reg,
    "Random Forest": rf_clf,
    "XGBoost": xgb_clf
}

results = {}

# Train and evaluate models
for name, model in models.items():
    print(f"Training {name}...")
    if name == "XGBoost":
        model.fit(X_train, y_train_encoded)
        y_pred_encoded = model.predict(X_test)
        y_pred = label_encoder.inverse_transform(y_pred_encoded) # Decode predictions
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)


    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted') # Use weighted for multiclass

    results[name] = {"Accuracy": accuracy, "F1 Score": f1}
    print(f"{name} - Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}\n")

print("Model Evaluation Results:")
for name, metrics in results.items():
    print(f"{name}: Accuracy={metrics['Accuracy']:.4f}, F1 Score={metrics['F1 Score']:.4f}")
```

Use hyperparameter tuning to optimize the best-performing model.

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from scipy.stats import uniform, randint
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# Ensure X_selected and y are available from previous steps
# Assuming X_selected contains the numerical features and y is the target variable

# Re-split the data using X_selected and y to ensure correct data is used
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)

# Initialize LabelEncoder and encode the target variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(100, 1000),
    'learning_rate': uniform(0.01, 0.3),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'gamma': uniform(0, 0.5),
}

# Initialize XGBoost classifier
# Set enable_categorical=False as we are using only numerical features
xgb_clf = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss', enable_categorical=False)


# Initialize RandomizedSearchCV
# Using F1 score (weighted) as the scoring metric
random_search = RandomizedSearchCV(
    xgb_clf, param_distributions=param_dist, n_iter=50, # You can adjust n_iter
    scoring='f1_weighted', cv=3, verbose=1, random_state=42, n_jobs=-1
)

# Fit RandomizedSearchCV to the training data
random_search.fit(X_train, y_train_encoded)

print("Best hyperparameters found:")
print(random_search.best_params_)

# Evaluate the best model on the test data
best_xgb_model = random_search.best_estimator_
y_pred_encoded = best_xgb_model.predict(X_test)
y_pred = label_encoder.inverse_transform(y_pred_encoded) # Decode predictions

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"\nBest XGBoost Model Performance on Test Set:")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score (Weighted): {f1:.4f}")
```
