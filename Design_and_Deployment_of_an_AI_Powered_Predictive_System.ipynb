## Address outliers

### Subtask:
Based on visualizations, decide on a strategy to handle outliers (e.g., capping, transformation, removal).


**Reasoning**:
Based on the box plots from the previous step, the columns 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', and 'thal' show potential outliers. I will implement outlier handling for these columns using capping at the 5th and 95th percentiles, as this is a reasonable approach for these features to mitigate the influence of extreme values without removing data points that might still hold relevant information.



```python
# Identify columns with potential outliers based on box plots:
outlier_cols = ['trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal']

# Implement capping at 5th and 95th percentiles for each identified column
for col in outlier_cols:
    # Calculate 5th and 95th percentiles
    lower_bound = df[col].quantile(0.05)
    upper_bound = df[col].quantile(0.95)

    # Apply capping
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

    print(f"Capped outliers for column: {col}")

# Regenerate box plots for the modified columns to visually confirm the effect
for col in outlier_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col} after outlier handling')
    plt.xlabel(col)
    plt.show()
```

## Data preprocessing

### Subtask:
Clean and transform the data for machine learning. This may involve handling missing values, encoding categorical variables, scaling numerical features, and splitting the data into training and testing sets.


**Reasoning**:
Separate features and target, identify categorical and numerical columns, apply one-hot encoding and standard scaling, concatenate the results, and split into training and testing sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Separate features (X) and target variable (y)
X = df.drop('target', axis=1)
y = df['target']

# 2. Identify categorical and numerical features
# Based on the data dictionary and previous analysis:
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Ensure all identified columns exist in X
categorical_features = [col for col in categorical_features if col in X.columns]
numerical_features = [col for col in numerical_features if col in X.columns]

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore') # handle_unknown='ignore' to handle potential unseen categories in test set

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# 3. & 4. Apply one-hot encoding and standard scaling, and 5. Concatenate
# The ColumnTransformer with a pipeline will handle the concatenation automatically
X_processed = preprocessor.fit_transform(X)

# Convert the processed data back to a DataFrame (optional, but good for inspection)
# Get feature names after one-hot encoding
feature_names = numerical_features + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features))
X_processed_df = pd.DataFrame(X_processed, columns=feature_names)

print("Shape of preprocessed features:", X_processed.shape)
print("\nFirst 5 rows of preprocessed features (as DataFrame):")
display(X_processed_df.head())

# 6. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y) # Using stratify to maintain target distribution

print("\nShape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
model = LogisticRegression(random_state=42)

# Train the model using the training data
model.fit(X_train, y_train)

print("Model training completed.")
```

## Flask application development

### Subtask:
Create a Flask application to serve the model. This involves setting up routes, handling requests, and using the saved model for predictions.


**Reasoning**:
Import necessary libraries, load the saved model and preprocessor, initialize the Flask app, and define the home and predict routes as per the instructions. The preprocessor is crucial for transforming new input data consistently with the training data.



```python
from flask import Flask, request, jsonify, render_template
import joblib
import numpy as np
import pandas as pd

# Load the trained model
try:
    model = joblib.load('model.joblib')
    print("Model loaded successfully.")
except FileNotFoundError:
    print("Error: model.joblib not found. Please ensure the model saving step was completed.")
    model = None # Set model to None if loading fails

# Load the preprocessor
# Note: The preprocessor was not explicitly saved in previous steps.
# We need to recreate it using the same configuration as during training.
# This requires knowing the numerical and categorical features used.
# Based on previous code:
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# To fit the preprocessor correctly, we need the original X data or a representative sample.
# Since df is available from previous cells, we can use it.
# Assuming df is still available and contains the original data before splitting and preprocessing.
try:
    # Create X from the original df, excluding the target
    X_original = df.drop('target', axis=1)
    # Fit the preprocessor
    preprocessor.fit(X_original)
    print("Preprocessor fitted successfully.")
except NameError:
    print("Error: Original DataFrame 'df' not found. Cannot fit preprocessor.")
    preprocessor = None # Set preprocessor to None if df is not available

# Initialize Flask app
app = Flask(__name__)

# Define home route
@app.route('/')
def home():
    # This will render an HTML file named 'index.html'
    # We will create this file in the next subtask
    return render_template('index.html')

# Define predict route
@app.route('/predict', methods=['POST'])
def predict():
    if model is None or preprocessor is None:
        return jsonify({'error': 'Model or preprocessor not loaded/fitted.'}), 500

    try:
        # Get input data from the request
        data = request.get_json(force=True) # Assuming input is JSON

        # Convert input data to DataFrame
        # Ensure column order matches the original feature order
        input_df = pd.DataFrame([data])

        # Apply the same preprocessing steps
        # Need to ensure input_df has the same columns as X_original and in the same order
        # This is a critical step for the preprocessor to work correctly
        # Let's ensure the input_df has all the original columns, even if some are missing in the input JSON (fill with default/NaN if necessary, depends on how the form is built)
        # For simplicity, let's assume the input JSON contains all necessary features in the correct order.
        # A more robust approach would involve explicit column mapping and handling missing keys.

        # Reorder columns to match the preprocessor's expected order
        expected_columns = numerical_features + categorical_features
        # Handle potential missing columns in input_df by adding them with NaN or a default
        for col in expected_columns:
            if col not in input_df.columns:
                input_df[col] = np.nan # Or a suitable default value

        input_df = input_df[expected_columns]


        # Transform the input data using the fitted preprocessor
        input_processed = preprocessor.transform(input_df)


        # Make prediction
        prediction = model.predict(input_processed)
        prediction_proba = model.predict_proba(input_processed) # Get probabilities

        # Convert prediction to readable format
        result = 'Disease' if prediction[0] == 1 else 'No Disease'

        # Return prediction as JSON response
        return jsonify({
            'prediction': result,
            'probability_disease': prediction_proba[0][1],
            'probability_no_disease': prediction_proba[0][0]
        })

    except Exception as e:
        return jsonify({'error': str(e)}), 400

# Standard Flask run command
# if __name__ == '__main__':
#     # In Colab, you typically don't run Flask directly like this.
#     # You would use ngrok or a similar tool to expose the port.
#     # For local development outside Colab, this line is used.
#     app.run(debug=True)
```

```python
%%writefile app.py
from flask import Flask, request, jsonify, render_template
import joblib
import numpy as np
import pandas as pd
import os

# Load the trained model
try:
    model = joblib.load('model.joblib')
    print("Model loaded successfully.")
except FileNotFoundError:
    print("Error: model.joblib not found. Please ensure the model saving step was completed.")
    model = None # Set model to None if loading fails

# Load the preprocessor
# Note: The preprocessor was not explicitly saved in previous steps.
# We need to recreate it using the same configuration as during training.
# This requires knowing the numerical and categorical features used.
# Based on previous code:
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# To fit the preprocessor correctly, we need the original X data or a representative sample.
# Assuming df is still available from previous cells.
# In a real deployment, you would save and load the fitted preprocessor.
# For this example, we'll refit it using the original data loaded from the file.
try:
    # Load original data again to fit the preprocessor
    df_original = pd.read_excel('heart.xlsx')
    # Remove duplicates if they were removed during initial loading
    if df_original.duplicated().sum() > 0:
        df_original.drop_duplicates(inplace=True)

    # Create X from the original df, excluding the target
    X_original = df_original.drop('target', axis=1)
    # Fit the preprocessor
    preprocessor.fit(X_original)
    print("Preprocessor fitted successfully.")
except FileNotFoundError:
    print("Error: heart.xlsx not found. Cannot fit preprocessor.")
    preprocessor = None # Set preprocessor to None if data file is not available
except NameError:
     print("Error: Could not access or load original DataFrame 'df'. Cannot fit preprocessor.")
     preprocessor = None

# Initialize Flask app
app = Flask(__name__)

# Define home route
@app.route('/')
def home():
    return render_template('index.html')

# Define predict route
@app.route('/predict', methods=['POST'])
def predict():
    if model is None or preprocessor is None:
        return jsonify({'error': 'Model or preprocessor not loaded/fitted.'}), 500

    try:
        # Get input data from the request
        data = request.get_json(force=True) # Assuming input is JSON

        # Convert input data to DataFrame
        # Ensure column order matches the original feature order used for preprocessing
        # Create a dictionary with all expected columns, filling with a default (e.g., 0 or NaN)
        # and then update with provided data. This handles cases where not all features
        # are provided in the input JSON (though the HTML form should provide all).
        expected_columns = numerical_features + categorical_features
        input_data_dict = {col: [np.nan] for col in expected_columns} # Use NaN or a suitable default

        # Update with actual data from the request
        for key, value in data.items():
             if key in input_data_dict:
                 # Convert data types as necessary based on original features
                 if key in numerical_features or key == 'oldpeak': # oldpeak is float
                     input_data_dict[key][0] = float(value) if key == 'oldpeak' else int(value)
                 else: # Categorical features (should be int)
                     input_data_dict[key][0] = int(value)
             # else: ignore unexpected keys in input data

        input_df = pd.DataFrame(input_data_dict)


        # Reorder columns to match the preprocessor's expected order just in case
        input_df = input_df[expected_columns]

        # Transform the input data using the fitted preprocessor
        input_processed = preprocessor.transform(input_df)

        # Make prediction
        prediction = model.predict(input_processed)
        prediction_proba = model.predict_proba(input_processed) # Get probabilities

        # Convert prediction to readable format
        result = 'Disease' if prediction[0] == 1 else 'No Disease'

        # Return prediction as JSON response
        return jsonify({
            'prediction': result,
            'probability_disease': float(prediction_proba[0][1]), # Convert numpy float to standard float
            'probability_no_disease': float(prediction_proba[0][0])
        })

    except Exception as e:
        # Log the error for debugging
        print(f"Prediction error: {e}")
        return jsonify({'error': str(e)}), 400

# To run the Flask app using pyngrok in Colab:
# You'll need to run this in a separate cell after writing the app.py file.
# from pyngrok import ngrok
#
# # Terminate any previous ngrok tunnels
# ngrok.kill()
#
# # Start a new ngrok tunnel
# # The default Flask port is 5000
# public_url = ngrok.connect(5000)
# print(f" * Ngrok tunnel established at {public_url}")
#
# # Run the Flask app
# # Use threaded=True to allow ngrok and Flask to run concurrently
# # In a production environment, use a WSGI server like Gunicorn
# # app.run(port=5000, threaded=True)
```

### Subtask:
Clean and transform the data for machine learning. This may involve handling missing values, encoding categorical variables, scaling numerical features, and splitting the data into training and testing sets.

Separate features and target, identify categorical and numerical columns, apply one-hot encoding and standard scaling, concatenate the results, and split into training and testing sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Separate features (X) and target variable (y)
X = df.drop('target', axis=1)
y = df['target']

# 2. Identify categorical and numerical features
# Based on the data dictionary and previous analysis:
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

# Ensure all identified columns exist in X
categorical_features = [col for col in categorical_features if col in X.columns]
numerical_features = [col for col in numerical_features if col in X.columns]

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore') # handle_unknown='ignore' to handle potential unseen categories in test set

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# 3. & 4. Apply one-hot encoding and standard scaling, and 5. Concatenate
# The ColumnTransformer with a pipeline will handle the concatenation automatically
X_processed = preprocessor.fit_transform(X)

# Convert the processed data back to a DataFrame (optional, but good for inspection)
# Get feature names after one-hot encoding
feature_names = numerical_features + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features))
X_processed_df = pd.DataFrame(X_processed, columns=feature_names)

print("Shape of preprocessed features:", X_processed.shape)
print("\nFirst 5 rows of preprocessed features (as DataFrame):")
display(X_processed_df.head())

# 6. Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y) # Using stratify to maintain target distribution

print("\nShape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from flask import Flask, request, jsonify, render_template
import joblib
import numpy as np
import pandas as pd

# Load the trained model
try:
    model = joblib.load('model.joblib')
    print("Model loaded successfully.")
except FileNotFoundError:
    print("Error: model.joblib not found. Please ensure the model saving step was completed.")
    model = None # Set model to None if loading fails

# Load the preprocessor
# Note: The preprocessor was not explicitly saved in previous steps.
# We need to recreate it using the same configuration as during training.
# This requires knowing the numerical and categorical features used.
# Based on previous code:
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore') # handle_unknown='ignore' to handle potential unseen categories in test set

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# To fit the preprocessor correctly, we need the original X data or a representative sample.
# Since df is available from previous cells, we can use it.
# Assuming df is still available and contains the original data before splitting and preprocessing.
try:
    # Create X from the original df, excluding the target
    X_original = df.drop('target', axis=1)
    # Fit the preprocessor
    preprocessor.fit(X_original)
    print("Preprocessor fitted successfully.")
except NameError:
    print("Error: Original DataFrame 'df' not found. Cannot fit preprocessor.")
    preprocessor = None # Set preprocessor to None if df is not available

# Initialize Flask app
app = Flask(__name__)

# Define home route
@app.route('/')
def home():
    # This will render an HTML file named 'index.html'
    # We will create this file in the next subtask
    return render_template('index.html')

# Define predict route
@app.route('/predict', methods=['POST'])
def predict():
    if model is None or preprocessor is None:
        return jsonify({'error': 'Model or preprocessor not loaded/fitted.'}), 500

    try:
        # Get input data from the request
        data = request.get_json(force=True) # Assuming input is JSON

        # Convert input data to DataFrame
        # Ensure column order matches the original feature order
        input_df = pd.DataFrame([data])

        # Apply the same preprocessing steps
        # Need to ensure input_df has the same columns as X_original and in the same order
        # This is a critical step for the preprocessor to work correctly
        # Let's ensure the input_df has all the original columns, even if some are missing in the input JSON (fill with default/NaN if necessary, depends on how the form is built)
        # For simplicity, let's assume the input JSON contains all necessary features in the correct order.
        # A more robust approach would involve explicit column mapping and handling missing keys.

        # Reorder columns to match the preprocessor's expected order
        expected_columns = numerical_features + categorical_features
        # Handle potential missing columns in input_df by adding them with NaN or a default
        for col in expected_columns:
            if col not in input_df.columns:
                input_df[col] = np.nan # Or a suitable default value

        input_df = input_df[expected_columns]


        # Transform the input data using the fitted preprocessor
        input_processed = preprocessor.transform(input_df)


        # Make prediction
        prediction = model.predict(input_processed)
        prediction_proba = model.predict_proba(input_processed) # Get probabilities

        # Convert prediction to readable format
        result = 'Disease' if prediction[0] == 1 else 'No Disease'

        # Return prediction as JSON response
        return jsonify({
            'prediction': result,
            'probability_disease': prediction_proba[0][1],
            'probability_no_disease': prediction_proba[0][0]
        })

    except Exception as e:
        return jsonify({'error': str(e)}), 400

# Standard Flask run command
# if __name__ == '__main__':
#     # In Colab, you typically don't run Flask directly like this.
#     # You would use ngrok or a similar tool to expose the port.
#     # For local development outside Colab, this line is used.
#     app.run(debug=True)
```
